```python
def get_infos(url):  
    chrome=get_chrome(url)
    name=Select(chrome.find_element_by_name('ctl00$ContentPlaceHolder1$SelGame'))
    date=Select(chrome.find_element_by_name('ctl00$ContentPlaceHolder1$SelBStatus'))
    
    name.select_by_value('ROM')
    date.select_by_value('2022/03')
    time.sleep(2)
    
    soup=BeautifulSoup(chrome.page_source,'lxml')
    counts=soup.find('div',class_='pagination').find_all('div',class_='pagination_select')
    
    infos_all=[]
    for count in counts:
        num=eval(count.find_all('option')[-1].text.strip())
        
        infos=[]
        page=0
        while page<num:
            time.sleep(2)
            soup=BeautifulSoup(chrome.page_source,'lxml')
            datas=soup.find('div',class_='l-records-content').find_all('div',class_='l-records-content_list logItem')
            
            for data in datas:
                
                player_name=data.find('li',class_='logItem-info_id').text.strip()
                server=data.find('li',class_='logItem-info_serve').text.strip()
                info_time=data.find('li',class_='logItem-info_time').text.strip()
                reason=data.find('ul',class_='logItem-suspend').text.strip()
                result=data.find('li',class_='logItem-suspend_result').text.strip()

                infos.append([player_name,server,info_time,reason,result])
            
            chrome.find_element_by_link_text(u'下一頁').click()
            page = page + 1
            time.sleep(2)

        infos_all+=infos
        
    return infos_all
```
